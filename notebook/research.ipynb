```python
import pandas as pd
import numpy as np  
```

```python
books = pd.read_csv('BX-Books.csv', sep=';', encoding='latin-1', on_bad_lines='skip')
users = pd.read_csv('BX-Users.csv', sep=';', encoding='latin-1', on_bad_lines='skip')
ratings = pd.read_csv('BX-Book-Ratings.csv', sep=';', encoding='latin-1', on_bad_lines='skip')
```

```python
# let's rename some wierd columns name

books.rename(columns={'Book-Title' : 'Title',
                      'Book-Author' : 'Author',
                      'Year-Of-Publication': 'Year',
                      'Publisher' : 'Publisher',
                      'Image-URL-L' : 'Image_url'}, inplace=True)
```

```python
ratings.rename(columns={'Book-Rating' : 'Rating'}, inplace=True)
```

```python
ratings.head()
```

```python
ratings['User-ID'].value_counts()
```

```python
# let's store users who had at least rated more than 200 books 

x = ratings['User-ID'].value_counts() > 200
```

```python
x
```

```python
x[x].shape
```

```python
y = x[x].index
```

```python
y
```

```python
ratings = ratings[ratings['User-ID'].isin(y)]
```

```python
ratings.head()
```

```python
# now join ratings with books 

books_with_ratings = ratings.merge(books, on='ISBN')
```

```python
books_with_ratings.head()
```

```python
number_of_ratings = books_with_ratings.groupby('Title')['Rating'].count().reset_index()
```

```python
number_of_ratings.head()
```

```python
number_of_ratings.rename(columns={'Rating' : 'num_of_Rating'}, inplace= True)
```

```python
number_of_ratings.head()
```

```python
final_rating = books_with_ratings.merge(number_of_ratings, on='Title')
```

```python
final_rating.head()
```

```python
# Lets Take Those Books Which Got At Least 50 ratings of users

final_rating = final_rating[final_rating['num_of_Rating'] >= 50]
```

```python
final_rating.head()
```

```python
final_rating.shape
```

```python
# lets drop duplicates value in columns User-ID and Title

final_rating.drop_duplicates(['User-ID', 'Title'], inplace=True)
```

```python
final_rating.shape
```

```python
final_rating.isnull().sum()
```

```python
# lets create a pivot table 

books_pivot = final_rating.pivot_table(columns='User-ID', index='Title', values='Rating')
```

```python
books_pivot
```

```python
books_pivot.shape
```

```python
books_pivot.fillna(0, inplace=True)
```

```python
books_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
books_sparse = csr_matrix(books_pivot)
```

```python
books_sparse
```

```python
type(books_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(books_sparse)
```

```python
books_pivot.iloc[237, :]
```

```python
distances, suggestions = model.kneighbors(books_pivot.iloc[237, :].values.reshape(1, -1), n_neighbors=6)
```

Testing Model :- 

```python
def recommend_book(book_name):
    book_id = np.where(books_pivot.index == book_name)[0][0]
    distances, suggestions = model.kneighbors(books_pivot.iloc[book_id, :].values.reshape(1, -1), n_neighbors=6)

    for i in range(len(suggestions)):
        books = books_pivot.index[suggestions[i]]
        for j in books:
            if j == book_name:
                print(f"You Searched Book {book_name}\n")
                print("The Suggestions Books Are:-\n")
            else:
                print(j)
```

```python
book_name = '84 Charing Cross Road'
```

```python
recommend_book(book_name)
```
